### QUIZ] 6개의 건물명을 리스트로 받아 도로명주소를 구한 뒤 데이터프레임으로 만들어 csv파일(addr.csv)로 만들기

```python
bldg_list = ['한국경제신문사','서울특별시청','중구청','종로구청','효창공원']
```

```python
road_url ='https://www.juso.go.kr/addrlink/addrLinkApi.do'
url_list=[]
for i in range(len(bldg_list)):
    url_list.append( f'{road_url}?confmKey={road_key}&currentPage=1&countPerPage=10&keyword={bldg_list[i]}&resultType=json' )
url_list
```

```python
result_list = []
for i in range(len(bldg_list)):
    result_list.append( requests.get(url_list[i]).json() ) 
result_list
```

```python
addr_list = []
for i in range(len(bldg_list)):
    for k in range(len(result_list[i]['results']['juso'])):
        road_name = result_list[i]['results']['juso'][k]['roadAddr']
        addr_list.append(road_name)
addr_list
```

```python
data = {
    "건물명" : bldg_list,
    "주소명" : addr_list 
}
df = pd.DataFrame(data)
df
```

### SOL]

```python
addr_list = []
for bldg in bldg_list:
    url = f"{road_url}?confmKey={road_key}&currentPage=1&countPerPage=10&keyword={quote(bldg)}&resultType=json"
    result = requests.get(url).json()
    road_address = result['results']['juso'][0]['roadAddr']
    addr_list.append(road_address)
```

```python
addr_list
```

```python
df = pd.DataFrame({
    '건물명': bldg_list,
    '주소': addr_list
})
df
```

```python
df.to_csv('addr.csv', index=False, sep=',', encoding='utf8')
```
