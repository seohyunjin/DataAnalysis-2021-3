```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
## 서울시 구별 주유소 가격 정보 얻기
driver = webdriver.Chrome()
driver.get('https://www.opinet.co.kr/user/main/mainView.do')
```

```python
driver.find_element_by_css_selector('.ic_m1').click() # 해당클릭  # 위와 동시에 실행하면 작동되지않음
```

```python
# css_selector 표시법 : "#아이디명" 
region = driver.find_element_by_css_selector("#SIGUNGU_NM0") 
gu_list = region.find_elements_by_css_selector('option')
gu_list
```

```python
# xpath 표시법 : '//*[@id="아이디명"]'
region = driver.find_element_by_xpath('//*[@id="SIGUNGU_NM0"]') 
gu_list = region.find_elements_by_css_selector('option')
gu_list
```

```python
# 서울 자치구 리스트 가져오기
region = driver.find_element_by_xpath('//*[@id="SIGUNGU_NM0"]') 
gu_list = region.find_elements_by_css_selector('option')
gu_names = [gu.text for gu in gu_list]
del gu_names[0] # 칼럼명0열 삭제 
gu_names
```

```python
gu = gu_names[0]
region = driver.find_element_by_css_selector("#SIGUNGU_NM0")
region.send_keys(gu)
```

```python
#엑셀 저장 버튼 클릭
driver.find_element_by_xpath('//*[@id="glopopd_excel"]').click()
```

```python
for gu in gu_names:
    region = driver.find_element_by_css_selector("#SIGUNGU_NM0")
    region.send_keys(gu)
    time.sleep(1)
    
    driver.find_element_by_xpath('//*[@id="glopopd_excel"]').click()  # 파일로 저자
    time.sleep(2) 
```

```python
driver.close()
```

## 구별 휘발유 가격에 대한 데이터 정리

```python
import pandas as pd
from glob import glob
```

```python
glob('C:\\Users\\CPB06GameN\\Downloads\\지역*.xls') 
# 해당 폴더의 '지역'으로 시작하는 모든 xls파일 가져옴
```

```python
file = glob('C:\\Users\\CPB06GameN\\Downloads\\지역*.xls')[0]
df = pd.read_excel(file, header= 2) #excel파일의 앞의 2줄은 건너띄고 가져오므로
df.head()
```

```python
raw = []
for file in glob('C:\\Users\\CPB06GameN\\Downloads\\지역*.xls'):
    df = pd.read_excel(file, header= 2)
    raw.append(df)
station_raw = pd.concat(raw) # 데이터 프레임 합치기 : concat 

station_raw.tail()
```

```python
station_raw.info() #  487 entries, 다 들어간 것 확인
```

```python
stations = pd.DataFrame({
    '상호' : station_raw.상호,
    '주소' : station_raw.주소,
    '가격' : station_raw.휘발유,
    '셀프' : station_raw.셀프여부,
    '상표': station_raw.상표 
})
stations.head()
```

```python
stations['구'] = [addr.split()[1] for addr in station_raw.주소]
stations.tail()
```

```python
stations['구'].unique()
```

```python
# - 없는 것만 뽑아내는 filtering
stations = stations[stations['가격'] != '-']
stations.info() # 474 
```

```python
stations['가격'] = pd.to_numeric(stations['가격'])
stations.info() # 가격이 int 로바뀐것을 확인
```
