```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["axes.unicode_minus"] = False
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

## - Count Plot

```python
sns.countplot( x = 'class', data = titanic )
plt.title('타이타닉호의 각 클래스별 승객 수')
plt.show()
```

```python
sns.countplot(x="day", data=tips)
plt.title("요일별 팁을 준 횟수")
plt.show()
```

## - Pariplot

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.title("Iris Pair Plot, Hue로 꽃의 종을 시각화")
plt.show()
```

## Heatmap

```python
flights_passengers = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(8,6))
plt.title("연도, 월 별 승객수에 대한 Heatmap")
sns.heatmap(flights_passengers, annot=True, fmt="d", linewidths=1) #피봇데이터로 만들어야함
plt.show()
```

##  Barplot

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("요일 별, 전체 요금의 평균")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", hue='sex', data=tips)
plt.title("요일 별/성별 전체 요금의 평균")
plt.show()
```

## Boxplot

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("요일 별, 전체 요금의 Box Plot")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue='sex', data=tips)
plt.title("요일 별/성별 전체 요금의 Box Plot")
plt.show()
```

## stripplot

```python
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("요일 별 전체 팁의 Strip Plot")
plt.show()
```

```python
# jitter=False
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", hue="sex", data=tips, jitter=False) 
#jitter : 흩뿌리기(겹쳐진 점들에 효과를 주어 시각적으로 나타낼 수 있음)
plt.title("요일 별, 성별 전체 팁의 Strip Plot")
plt.legend(loc=1)
plt.show()
```

```python
# jitter=True
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", hue="sex", data=tips, jitter=True) 
#jitter : 흩뿌리기(겹쳐진 점들에 효과를 주어 시각적으로 나타낼 수 있음)
plt.title("요일 별, 성별 전체 팁의 Strip Plot")
plt.legend(loc=1)
plt.show()
```
