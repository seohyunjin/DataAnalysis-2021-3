# 여러개의 그림 그리기

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams["font.family"] = 'NanumBarunGothic'
plt.rcParams["axes.unicode_minus"] = False
```

## - figure 객체

```python
np.random.seed(2021)
plt.figure(figsize=(10,2)) #그림의 크기 설정
plt.title('figure size:(10, 2)')
plt.plot(np.random.randn(100))
plt.grid()
plt.show()
```

## - Axes 객체와 subplot

### - subplot
figure 그림

\[1] \[2]

\[3] \[4]

\[1]위치에 그림그리고 싶다면,  subplot (2,2,1)

\[2]위치에 그림그리고 싶다면,  subplot (2,2,2)


```python
x1 = np.linspace(0.,5.)
x2 = np.linspace(0.,2.)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
ax1 = plt.subplot(2,1,1)
plt.title('2 Subplots') # title 코드 위치가 첫번째 그래프사이에 넣어주어야 정위치에 생김 
ax1.plot(x1, y1, 'yo-') # 'yo-' ; yellow + o + - 설정 
ax1.set_ylabel('Damped oscillation') # plt에서는 ylabel이나, ax1에서는 set_ylabel

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-') #r.- ; red +. +  -
plt.ylabel('Undamped')

plt.xlabel('time (sec)')
plt.show()
```

```python
ax1 = plt.subplot(211)
plt.title('2 Subplots') # title 코드 위치가 첫번째 그래프사이에 넣어주어야 정위치에 생김 
ax1.plot(x1, y1, 'yo-') # 'yo-' ; yellow + o + - 설정 
ax1.set_ylabel('Damped oscillation') # plt에서는 ylabel이나, ax1에서는 set_ylabel
plt.grid() # grid해줄때 각각 해주어야함

ax2 = plt.subplot(212)
plt.plot(x2, y2, 'r.-') #r.- ; red +. +  -
plt.ylabel('Undamped')
plt.xlabel('time (sec)')
plt.grid() # grid해줄때 각각 해주어야함

# plt.tight_layout()
plt.show()
```

### - subplots
subplots 명령으로 복수의 Axes 객체를 동시에 생성

이때는 2차원 ndarray 형태로 Axes 객체가 반환

```python
# 위와 같은 결과 출력 위와 같은 결과 출력
fig,axes = plt.subplots(2,2) # 배경설정

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# 반복문써서 그래프 그리는 것 권장
np.random.seed(0)

fig,axes = plt.subplots(2,2) # 배경설정
for i in range(4):
    ax = axes[i//2, i%2] # 2,2 figure로 그리므로, 행인덱스: i//2 , 열인덱스:i%2
    ax.plot(np.random.rand(5))
    ax.set_title(f'axes{i+1}')
    ax.grid()

plt.tight_layout()
plt.show()
```
