# Pandas의 시각화 기능
Pandas의 시리즈나 데이터프레임은 plot이라는 시각화 메서드를 내장

plot은 matplotlib를 내부에서 임포트하여 사용

```python
np.random.seed(2021)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2021', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot()
plt.title("Pandas의 Plot메소드 사용 예")
plt.xlabel("시간")
plt.ylabel("Data")
plt.grid()
plt.show()
```

## plot 메서드 - kind 인수

- bar

- pie

- hist

- kde

- box

- scatter

- area

### - bar

```python
iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.title("Iris 데이터의 Bar Plot")
plt.xlabel("Data")
plt.ylabel("각 Feature의 값")
plt.ylim(0, 7)
plt.show()
```

```python
df2 = iris.groupby('species').mean()
df2.columns.name = "feature"
df2
```

```python
df2.plot.bar()
plt.title("각 종의 Feature별 평균")
plt.xlabel("평균")
plt.ylabel("단위(cm))")
plt.ylim(0, 8)
plt.grid()
plt.show()
```

```python
df2.T.plot.bar() # 그래프 사이즈 조정하고 싶다면 bar()의 괄호안에 figsize= (10, 8) 속성추가
plt.title("각 종의 Feature별 평균")
plt.xlabel("평균")
plt.ylabel("단위(cm))")
plt.ylim(0, 8)
plt.grid()
plt.show()
```

### - pie

```python
df3 = titanic.pclass.value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title("선실별 승객 수 비율")
plt.axis('equal')
plt.show()
```

### - box

```python
iris.plot.box()
plt.title("각 Feature 값들에 대한 Box Plot")
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.show()
```

```python
iris.boxplot(by='species', figsize= (10, 8))
plt.tight_layout(pad=2, h_pad=1)
plt.title("각 Feature의 종 별 데이터에 대한 Box Plot")
plt.show()
```
