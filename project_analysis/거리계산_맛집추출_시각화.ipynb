## 한 대여소에 대하여 거리계산

### - 37번 대여소에 대하여

```python
data_s[37]
```

```python
# 
# wanted_r = [] # only 위경도
wanted_total =[] # 지도에 나타낼 정보 + 위경도
s_inf_latitude = data_s[37][3] # 위도
s_inf_longtitude = data_s[37][4] #경도
for k in range(len(data_g)):
    g_inf_latitude = data_g[k][5] # 위도
    g_inf_longtitude = data_g[k][4] #경도

    start = (float(s_inf_latitude), float(s_inf_longtitude)) # 대여소(기준) 위치
    goal = (float(g_inf_latitude), float(g_inf_longtitude)) # 별점3점이상 음식점 위치
    distance = haversine(start, goal, unit = 'km') # 두 관측치 사이 거리
    
    if distance <= 1.0:
        # wanted_r.append(goal)
        wanted_total.append(data_g[k])

wanted_total[0:3]
```

```python
r_lat_list =[]# 위도
r_lag_list =[]# 경도
r_name_list=[] # 상점명
r_index_list=[] # 인덱스번호
r_addr_list=[] # 주소
r_star_list=[] # 별점
for i in range(len(wanted_r)):
    r_index_list.append(wanted_total[i][0]) # 인덱스번호
    r_name_list.append(wanted_total[i][1]) # 상점명
    r_addr_list.append(wanted_total[i][2]) # 주소
    r_star_list.append(wanted_total[i][3]) # 별점
    r_lat_list.append(wanted_total[i][5]) #위도리스트
    r_lag_list.append(wanted_total[i][4]) #경도리스트
```
